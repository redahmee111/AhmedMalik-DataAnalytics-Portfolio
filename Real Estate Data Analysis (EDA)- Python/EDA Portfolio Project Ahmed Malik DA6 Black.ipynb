## 4. Data Preprocessing
- Converted `price` to numeric PKR
- Converted `area` to `area_sqft` uniform unit
- Extracted `beds` and `baths` where possible
- Created `price_per_sqft` column

```python
def clean_price(value):
    if pd.isna(value):
        return np.nan
    s = str(value).lower().replace(',', ' ').strip()
    s = s.replace('â‚¨','').replace('rs.','').replace('rs','').replace('pkr','')
    if 'crore' in s:
        m = re.search(r'([\d\.]+)', s)
        return float(m.group(1)) * 1e7 if m else np.nan
    if 'lakh' in s or 'lac' in s:
        m = re.search(r'([\d\.]+)', s)
        return float(m.group(1)) * 1e5 if m else np.nan
    m = re.search(r'([\d\.]+)', s)
    return float(m.group(1)) if m else np.nan

if 'price' in df.columns:
    df['price_num'] = df['price'].apply(clean_price)
else:
    df['price_num'] = np.nan

print('Price conversion complete. Sample:')
df[['price','price_num']].head(10)
```

```python
def area_to_sqft(text):
    if pd.isna(text): return np.nan
    s = str(text).lower().replace(',', ' ').strip()
    m = re.search(r'([\d\.]+)', s)
    if not m:
        return np.nan
    val = float(m.group(1))
    if 'marla' in s: return val * 272.25
    if 'kanal' in s: return val * 5445
    if 'sqft' in s or 'sq. ft' in s or 'square feet' in s: return val
    if 'sqm' in s or 'm2' in s or 'mÂ²' in s: return val * 10.7639
    if 'yard' in s or 'sqyd' in s or 'square yard' in s: return val * 9
    return val

if 'area' in df.columns:
    df['area_raw'] = df['area'].astype(str)
    df['area_sqft'] = df['area'].apply(area_to_sqft)
else:
    df['area_raw'] = ''
    df['area_sqft'] = np.nan

df[['area','area_sqft']].head(10)
```

```python
beds_candidates = [c for c in df.columns if 'bed' in c.lower()]
baths_candidates = [c for c in df.columns if 'bath' in c.lower()]

if beds_candidates:
    df['beds'] = pd.to_numeric(df[beds_candidates[0]], errors='coerce')
else:
    df['beds'] = df['title'].astype(str).str.extract(r'(\d+)\s*bed', expand=False).astype(float)

if baths_candidates:
    df['baths'] = pd.to_numeric(df[baths_candidates[0]], errors='coerce')
else:
    df['baths'] = df['description'].astype(str).str.extract(r'(\d+)\s*bath', expand=False).astype(float)

df[['title','beds','baths']].head(8)
```

```python
if 'city' in df.columns:
    df['city'] = df['city'].astype(str).str.strip().str.title()
else:
    df['city'] = 'Unknown'

if 'property_type' in df.columns:
    df['property_type'] = df['property_type'].astype(str).str.strip().str.title()
else:
    df['property_type'] = 'Unknown'
```

```python
df['price_per_sqft'] = df['price_num'] / df['area_sqft']
date_cols = [c for c in df.columns if 'date' in c.lower() or 'posted' in c.lower()]
if date_cols:
    try:
        df['date_posted'] = pd.to_datetime(df[date_cols[0]], errors='coerce')
        df['listing_age_days'] = (pd.Timestamp.today() - df['date_posted']).dt.days
    except Exception as e:
        df['listing_age_days'] = np.nan
else:
    df['listing_age_days'] = np.nan

df[['price_num','area_sqft','price_per_sqft','listing_age_days']].head()
```
